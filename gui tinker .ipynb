```python
root = tk.Tk()

root.title("Learning Radiobutton")
root.geometry("400x400")

v = tk.IntVar()

L1 = Label(root, text= "Choose a Programming Language", justify=tk.LEFT, padx = 20)
L1.pack()


B1 = Radiobutton(root, text="Python" , padx = 20,  variable=v, value=1)
B1.pack()


B2 = Radiobutton(root, text="Javascript" , padx = 20, variable=v, value=2)
B2.pack()


B3 = Radiobutton(root, text="Oracle" , padx = 20, variable=v, value=3)
B3.pack()


root.mainloop()
```

```python
root = tk.Tk()

root.title("Learning Checkbutton")
root.geometry("400x400")
root.configure(bg="yellow")

L1 = Label(root, text= "Choose a Fruit Name", justify=tk.LEFT, padx = 20)
L1.grid(row=0,column=1)

v1 = IntVar()

B1= Checkbutton(root, text="Apple", variable=v1)
B1.grid(row=1, column=1)

v2 = IntVar()

B2 = Checkbutton(root, text="Cherry", variable=v2)
B2.grid(row=2, column=1)

v3 = IntVar()

B3 = Checkbutton(root, text="berry", variable=v3)
B3.grid(row=3, column=1)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import *

win = tk.Tk()

mylable = tk.Label(win, text="Guess the number 10", bg="pink")
mylable.grid(row=0, column= 1)

def click_me1():
    B1.configure(text="looser")
    mylable.configure(text="20")
    
def click_me2():
    B2.configure(text="winner")
    mylable.configure(text="10")
    
def click_me3():
    B3.configure(text="looser")
    mylable.configure(text="30")

def click_me4():
    B4.configure(text="looser")
    mylable.configure(text="40")
    
    
B1 = Button(win, text="guess 10", bg="yellow", command=click_me1)
B1.grid(row=1, column=1)

B2 = Button(win, text="guess 10", bg="yellow", command=click_me2)
B2.grid(row=2, column=1)

B3 = Button(win, text="guess 10", bg="yellow", command=click_me3)
B3.grid(row=3, column=1)

B4 = Button(win, text="guess 10", bg="yellow", command=click_me4)
B4.grid(row=4, column=1)


win.mainloop()
```
